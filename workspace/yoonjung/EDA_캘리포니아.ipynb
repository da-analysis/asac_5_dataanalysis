## rating

```
%sql
-- avg_rating 분포  
-- 

select avg_rating, count(*) as cnt from hive_metastore.silver.california_meta
group by avg_rating
order by cnt desc;
```

```
%sql
-- avg_rating 분포
-- 평균 평점 1인 gmap 개수가 4071개 

SELECT 
    CASE 
        WHEN avg_rating >= 1.0 AND avg_rating < 1.5 THEN '1.0-1.4'
        WHEN avg_rating >= 1.5 AND avg_rating < 2.0 THEN '1.5-1.9'
        WHEN avg_rating >= 2.0 AND avg_rating < 2.5 THEN '2.0-2.4'
        WHEN avg_rating >= 2.5 AND avg_rating < 3.0 THEN '2.5-2.9'
        WHEN avg_rating >= 3.0 AND avg_rating < 3.5 THEN '3.0-3.4'
        WHEN avg_rating >= 3.5 AND avg_rating < 4.0 THEN '3.5-3.9'
        WHEN avg_rating >= 4.0 AND avg_rating < 4.5 THEN '4.0-4.5'           
        WHEN avg_rating >= 4.5 AND avg_rating <= 5 THEN '4.5-5.0'                          
    END as rating_range,
    COUNT(*) as cnt
FROM hive_metastore.silver.california_meta
GROUP BY 
    CASE 
        WHEN avg_rating >= 1.0 AND avg_rating < 1.5 THEN '1.0-1.4'
        WHEN avg_rating >= 1.5 AND avg_rating < 2.0 THEN '1.5-1.9'
        WHEN avg_rating >= 2.0 AND avg_rating < 2.5 THEN '2.0-2.4'
        WHEN avg_rating >= 2.5 AND avg_rating < 3.0 THEN '2.5-2.9'
        WHEN avg_rating >= 3.0 AND avg_rating < 3.5 THEN '3.0-3.4'
        WHEN avg_rating >= 3.5 AND avg_rating < 4.0 THEN '3.5-3.9'
        WHEN avg_rating >= 4.0 AND avg_rating < 4.5 THEN '4.0-4.5'           
        WHEN avg_rating >= 4.5 AND avg_rating <= 5 THEN '4.5-5.0'                  
    END
ORDER BY rating_range;
```

# 실버 데이터 EDA_리뷰

## time

```
%sql

SELECT 
    MONTH(time) AS month, -- time 컬럼에서 월을 추출하여 month 컬럼 생성
    COUNT(*) AS review_count -- 월별 리뷰 개수를 카운트
FROM 
    hive_metastore.silver.review_data
WHERE 
    region = 'California'
GROUP BY 
    MONTH(time) -- month 컬럼을 기준으로 그룹화
ORDER BY 
    month; -- 월별로 정렬
```

```
%sql
SELECT 
    YEAR(time) AS year, -- time 컬럼에서 년도를 추출하여 year 컬럼 생성
    COUNT(*) AS review_count -- 년도별 리뷰 개수를 카운트
FROM 
    hive_metastore.silver.review_data
WHERE 
    region = 'California'
GROUP BY 
    YEAR(time) -- year 컬럼을 기준으로 그룹화
ORDER BY 
    year; -- 년도별로 정렬
```

## gmap_id

```
%sql
--  리뷰 개수별 리뷰 카운트 분포 확인 
-- cnt : gmap_id별 리뷰개수 
-- gmap_count : cnt별 개수 (cnt 리뷰가 1개인 gmap_id 가 17404개 )
-- sum : cnt별 총 리뷰개수 (cnt*review)
-- rate : 모든 리뷰 개수 6700만개에서 해당 리뷰개수가 차지하는 비율 


WITH base_data AS (
    SELECT 
        cnt, 
        COUNT(*) AS gmap_count
    FROM (
        SELECT gmap_id, COUNT(*) AS cnt 
        FROM hive_metastore.silver.california_review
        GROUP BY gmap_id
    ) AS grouped_data
    GROUP BY cnt
)
, sum_data AS (
    SELECT 
        cnt, 
        gmap_count, 
        cnt * gmap_count AS sum
    FROM base_data
)
, total_sum AS (
    SELECT 
        SUM(sum) AS total_sum_value 
    FROM sum_data
)
SELECT 
    sd.cnt, 
    sd.gmap_count, 
    sd.sum, 
    sd.sum / ts.total_sum_value AS rate
FROM 
    sum_data sd,
    total_sum ts
ORDER BY 
    sd.gmap_count DESC;
```

```
%sql

-- 범위 분포 표 
-- -> 리뷰 1개만 있는 gmap_id 수 : 17404 


SELECT 
    CASE
        WHEN cnt >= 1 AND cnt < 2 THEN '1'
        WHEN cnt >= 2 AND cnt < 3 THEN '2'
        WHEN cnt >= 3 AND cnt < 4 THEN '3'
        WHEN cnt >= 4 AND cnt < 5 THEN '4'
        WHEN cnt >= 5 AND cnt < 6 THEN '5'
        WHEN cnt >= 6 AND cnt < 10 THEN '6-9'
        WHEN cnt >= 10 AND cnt < 100 THEN '10-99'
        WHEN cnt >= 100 AND cnt < 1000 THEN '100-999'
        WHEN cnt >= 1000 AND cnt < 5000 THEN '1000-4999'
        ELSE '5000+'
    END AS review_count_range,
    COUNT(*) AS count
FROM 
    (
        SELECT gmap_id, COUNT(*) AS cnt
        FROM hive_metastore.silver.california_review
        GROUP BY gmap_id
    ) AS review_counts
GROUP BY 
    CASE
        WHEN cnt >= 1 AND cnt < 2 THEN '1'
        WHEN cnt >= 2 AND cnt < 3 THEN '2'
        WHEN cnt >= 3 AND cnt < 4 THEN '3'
        WHEN cnt >= 4 AND cnt < 5 THEN '4'
        WHEN cnt >= 5 AND cnt < 6 THEN '5'
        WHEN cnt >= 6 AND cnt < 10 THEN '6-9'
        WHEN cnt >= 10 AND cnt < 100 THEN '10-99'
        WHEN cnt >= 100 AND cnt < 1000 THEN '100-999'
        WHEN cnt >= 1000 AND cnt < 5000 THEN '1000-4999'
        ELSE '5000+'
    END
ORDER BY count DESC;
```

## user_id

```
%sql
--  User_id 기준 - 한 사용자가 몇 개의 gmap_id 장소를 갔는지
-- 장소 방문 개수별 분포 확인


select unique_gmap_count, count(*) as cnt
from 
  (select user_id, count(distinct gmap_id) as unique_gmap_count
  from hive_metastore.silver.review_data
  where region = 'California' and user_id is not null
  group by user_id)
group by unique_gmap_count
order by unique_gmap_count;
```

```
%sql
--  User_id 기준 - 한 사용자가 몇 개의 gmap_id 장소를 갔는지
-- 유저의 장소 방문 개수별 분포 확인


select unique_gmap_count, count(*) as cnt
from 
  (select user_id, count(distinct gmap_id) as unique_gmap_count
  from hive_metastore.silver.california_review
  where user_id is not null
  group by user_id)
group by unique_gmap_count
order by unique_gmap_count;
```

## text

```
%sql

-- 리뷰가 1개인 것의 text 에 null 값인 개수와 아닌 개수

SELECT 
    CASE 
        WHEN text = 'N/A' THEN 'N/A'
        ELSE 'Not N/A'
    END AS text_status,
    COUNT(*) AS count
FROM hive_metastore.silver.california_review
WHERE gmap_id IN (
    SELECT gmap_id
    FROM (
        SELECT gmap_id, COUNT(*) AS cnt
        FROM hive_metastore.silver.california_review
        GROUP BY gmap_id
        HAVING cnt = 1
    ) AS filtered_reviews
)
GROUP BY 
    CASE 
        WHEN text = 'N/A' THEN 'N/A'
        ELSE 'Not N/A'
    END;
```

```
%sql

-- 리뷰가 1개~5개인 것의 text 에 null 값인 개수와 아닌 개수

SELECT 
    CASE 
        WHEN text = 'N/A' THEN 'N/A'
        ELSE 'Not N/A'
    END AS text_status,
    COUNT(*) AS count
FROM hive_metastore.silver.california_review
WHERE gmap_id IN (
    SELECT gmap_id
    FROM (
        SELECT gmap_id, COUNT(*) AS cnt
        FROM hive_metastore.silver.california_review
        GROUP BY gmap_id
        HAVING cnt >=1 and cnt<=5
    ) AS filtered_reviews
)
GROUP BY 
    CASE 
        WHEN text = 'N/A' THEN 'N/A'
        ELSE 'Not N/A'
    END;
```

```
## 정리 - 데이터가 많아서 유용하지 않는 데이터는 컷하고 샘플링하여 연관규칙(co-count) 진행하고자 할 때 확인한 사항
# -> 리뷰개수가 1개 인 것 : gmap_id 17404개 , 총 리뷰개수 17,404 정도됨 / 리뷰 텍스트 null 값이 1/3
# -> 리뷰개수가 1-5 인 것 : gmap_id  68113개, 총 리뷰개수 188,726 정도됨 / 리뷰 텍스트 null 값이 1/2

#--> 텍스트 리뷰가 null (N/A)인 값이 총 리뷰개수의 반 정도 됨 
#--> 텍스트 null 값을 제외하여 3700만개 정도로 줄여서 사용할 수 있음
```
